```python
import pandas as pd
import numpy as np
import random
import string

# Fungsi untuk membuat nama acak
def generate_random_name(num_names=1):
    names = []
    for _ in range(num_names):
        # Menghasilkan nama dengan kombinasi huruf acak (5-10 karakter)
        name_length = random.randint(5, 10)
        name = ''.join(random.choices(string.ascii_uppercase, k=name_length))
        names.append(name)
    return names

# Fungsi untuk membuat dataset
def generate_dataset(num_samples=50):
    # Inisialisasi list untuk setiap fitur
    nilai = np.random.randint(50, 100, size=num_samples)  # Skor 50-100
    umur = np.random.randint(18, 25, size=num_samples)  # Usia 18-24 tahun
    penghasilan_ayah = np.random.randint(0, 10000000, size=num_samples)  # Rp 0 - 10 juta
    penghasilan_ibu = np.random.randint(0, 8000000, size=num_samples)  # Rp 0 - 8 juta
    saudara = np.random.randint(0, 6, size=num_samples)  # Jumlah saudara 0-5
    nama = generate_random_name(num_samples)  # Nama acak
    
    # Buat DataFrame
    data = {
        "Nama": nama,
        "Nilai": nilai,
        "Umur": umur,
        "Penghasilan_Ayah": penghasilan_ayah,
        "Penghasilan_Ibu": penghasilan_ibu,
        "Saudara": saudara
    }
    
    df = pd.DataFrame(data)
    return df

# Generate dataset dengan 50 sampel
dataset = generate_dataset(50)

# Tampilkan dataset
print(dataset)

# Simpan dataset ke file CSV jika diperlukan
dataset.to_csv("beasiswa_saw_dataset.csv", index=False)
```
